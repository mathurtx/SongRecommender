```python
import pandas as pd
from ggplot import *
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
user_product_rating = pd.read_csv("../../data/lastfm-dataset-1K/usageColabDataCastedv5.csv/part-r-00000-32c7094f-31b7-40c2-a19e-03d4733a7bc5.csv", header=None)
```

```python
user_proudct_rating = user_product_rating.rename(columns={0:'user',1:'product',2:'views'})
```

```python
user_product_rating.dtypes
```

```python
user_proudct_rating_1000 = user_product_rating[:1000]
```

```python
user_proudct_rating_1000 = user_proudct_rating_1000.rename(columns={0:'user',1:'product',2:'views'})
```

```python
user_proudct_rating_1000['user']= user_proudct_rating_1000['user'].astype(int)
```

```python
user_proudct_rating_1000['product'] = user_proudct_rating_1000['product'].astype(int)
```

```python
user_proudct_rating_1000['views'] = user_proudct_rating_1000['views'].apply(str)
```

```python
g=ggplot(user_proudct_rating_10000,aes(x='user',y='views',color='views')) +geom_point() +theme_bw()
```

```python
g.save("user_product_views.png")
```

```python
#data.groupby(by=['account_ID'])['purchases'].sum()
user_views = user_proudct_rating[['user','views']]
```

```python
user_views['total_views'] = user_views.groupby(['user'])['views'].sum()
```

```python
user_views.head(5)
```

```python
p=ggplot(user_views,aes(x='user',y='total_views')) +geom_point() +theme_bw()
p.save("user_views.png")
```

```python
track_views = user_proudct_rating[['product','views']]
track_views['track_total_views'] = track_views.groupby(['product'])['views'].sum()
pv=ggplot(track_views,aes(x='product',y='track_total_views')) +geom_point() +theme_bw()
pv.save("track_views.png")
```
